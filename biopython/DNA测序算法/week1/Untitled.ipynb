```python
def naive_with_rc(P,T):
    P_RC = reverseComplement(P)
    a1 = naive(P,T)
    c1 = len(a1)
    a2 = naive(P_RC,T)
    c2 = len(a2)
    sum = a1 + a2
    c3 = len(sum)
    
    if P == P_RC:
        return a1,c1
    return sum,c3
```

```python
oc, c = naive_with_rc(A,genome)
print(oc,'\n',c)
```

```python
B = Seq('AGTCGA')

oc, c = naive_with_rc(B,genome)
print(oc,'\n',c)
```

```python
def naive_2mm(p,t):
    occurrences = []
    for i in range(len(t) - len(p) + 1):  
        mmcount = 0
        for j in range(len(p)):  
            if t[i+j] != p[j]:  
                mmcount+=1
                if mmcount>2:
                    break
        if mmcount<=2 and j == (len(p)-1):
            occurrences.append(i) 
    return occurrences,'count is: '+str(len(occurrences))
```

```python
def findgc(reads):
    gc = [0]*65
    totals = [0]*65

    for i,read in enumerate(reads):
        for j in range(len(read)):
            if read[j] == ('C' or 'G'):
                gc[i]+=1

            totals[i]+=1

    print(gc)
    for i in range(len(gc)):
        if totals[i]>0:
            gc[i] /= float(totals[i])
        #else: gc[i]=0
    
    print(gc)
    print(totals)
    
    return gc
            
    
```
