```python
##frame 1
def ORF_v1(s1):
    start_idx, end1_idx, end2_idx, end3_idx = [],[],[],[]
    st, ed = None,None
    
    for i in range(0,len(s1),3):
        if i+2 <= len(s1)-1:
            if start == s1[i]+s1[i+1]+s1[i+2]:
                start_idx.append(i)
    
    for i in range(0,len(s1),3):
        if i+2 <= len(s1)-1:
            if end1 == s1[i]+s1[i+1]+s1[i+2]:
                end1_idx.append(i)
    
    for i in range(0,len(s1),3):
        if i+2 <= len(s1)-1:
            if end2 == s1[i]+s1[i+1]+s1[i+2]:
                end2_idx.append(i)
    
    for i in range(0,len(s1),3):
        if i+2 <= len(s1)-1:
            if end3 == s1[i]+s1[i+1]+s1[i+2]:
                end3_idx.append(i)
    
    
    print(start_idx)
    print(end1_idx,end2_idx,end3_idx)
    end_1 = sorted(end1_idx+end2_idx+end3_idx)
    print(end_1)
    
    if start_idx == None or end_1 == None:
        print("No ORF found.")
    
    t1 = time.perf_counter()
    m,maximum = 0,0
    for i in range(len(end_1)):
        for j in range(len(start_idx)):
            if i-1<0: 
                m = max(m,end_1[i]-start_idx[j])
            elif end_1[i] > start_idx[j] and end_1[i-1] < start_idx[j]:
                m = max(m,end_1[i]-start_idx[j])
            elif start_idx[j] > end_1[i]: 
                break
    
            if m > maximum:
                    maximum = m
                    st = start_idx[j]
                    ed = end_1[i]
    t2 = time.perf_counter()
    
    print('max OFR is:',maximum, '\nstart point is :',st,'\nend point is :',ed,'\ntime count:', t2-t1)
    
```
